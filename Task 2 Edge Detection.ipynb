```python
img=plt.imread("darksideofthemoon.jpg") #sample image taken
plt.imshow(img)
#kindly note tha tI have taken the image directly from my system's storage, via drag and drop method
```

```python
#applying sobel Filters
ver=[[-1,-2,-1],[0,0,0],[1,2,1]]
hor=[[-1,0,1],[-2,0,2],[-1,0,1]]
m,n,d=img.shape
print(m,n,d) #row, column, layers(for rgb)
```

```python
ver_img=np.zeros_like(img) #setting an array to zero(I found thsi function in a tutorial)
#otherwise we can simply comvert all teh following code into the code for working upon arrays
```

```python

for r in range(3,m-2):
  for c in range(3,n-2):
    local=img[r-1:r+2, c-1:c+2, 0]
    convoluted_output=ver*local          #taking 3*3 parts of the image matrix and taking its convoluted product with the filter matrix.
    ver_val=(convoluted_output.sum())/9
    ver_img[r,c]=[ver_val]*3
```

```python
plt.imshow(ver_img)
```

```python
hor_img=np.zeros_like(img)
for r in range(3,m-2):
  for c in range(3,n-2):
    local=img[r-1:r+2, c-1:c+2, 0]
    convoluted_output=hor*local
    hor_val=(convoluted_output.sum())/9
    hor_img[r,c]=[hor_val]*3
```

```python
plt.imshow(hor_img)
```

```python
edge_img=np.zeros_like(img)
for r in range(3,m-2):
  for c in range(3,n-2):
    local=img[r-1:r+2, c-1:c+2, 0]
    ver_convo=ver*local
    ver_val=ver_convo.sum()/5
    hor_convo=hor*local
    hor_val=hor_convo.sum()/5
    edge_val=(ver_val**2+hor_val**2)**0.5
    edge_img[r,c]=[edge_val]
```

```python
plt.imshow(edge_img)
```
